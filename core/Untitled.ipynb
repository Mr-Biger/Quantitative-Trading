```python
import numpy as np
import pandas as pd
```

```python
name = '600028'
datafram = pd.read_csv('../db/%s.csv'%name,sep=',',encoding='utf-8', index_col= 'date')
```

```python
datafram
```

```python
returns = (datafram['close'] - datafram['close'].shift(-1)) / datafram['close'].shift(-1)

returns
```

```python
max_range = returns.max() - returns.min()
mad = returns.mad()
var = returns.var()
std = returns.std()

print(max_range,mad,var,std)
```
